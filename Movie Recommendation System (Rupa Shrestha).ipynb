```python
#Importing the required packages

import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#Reading the data sets

df1 = pd.read_csv(r'E://E drive//Year 3 sem 1//AI//19031541 Rupa Shrestha Coursework-2//Movie Recommendation System//movies.csv')
df2 = pd.read_csv(r'E:/E drive/Year 3 sem 1/AI/19031541 Rupa Shrestha Coursework-2/Movie Recommendation System/ratings.csv')
```

```python
#merging dataset 1 and 2 based on movie id
df = df2.merge(df1, left_on='movieId', right_on='movieId', how='right')
#printing merged datasets
df
```

```python
#Removing columns which will not be used 

del df['timestamp']
del df['genres']
```

```python
df.head()
```

```python
# pivot ratings into movie features
user_movie_matrix = pd.pivot_table(df, values = 'rating', index='movieId', columns = 'userId')
user_movie_matrix.head()
```

```python
#relacing all NaN values with 0
user_movie_matrix = user_movie_matrix.fillna(0)
user_movie_matrix
```

Pandas dataframe.corr() is used to find the pairwise correlation of all columns in the dataframe. 

```python
#using collaborative filtering for calculating similarity score
# To find the correlation among the columns using pearson method

user_user_matrix = user_movie_matrix.corr(method='pearson', min_periods=1)
user_user_matrix.head()
```

```python
#Extarcing top 10 similar users for User2 by sorting them in descending order based on their similarties

user_user_matrix.loc[2].sort_values(ascending=False).head(11)
```

```python

#Converting the above data into a DF
df_2 = pd.DataFrame(user_user_matrix.loc[2].sort_values(ascending=False).head(11))
df_2 = df_2.reset_index()
df_2.columns = ['userId', 'similarity']

#Removing the user2 itself (A user will be always similar to iteself with a rating of 1)
df_2 = df_2.drop((df_2[df_2['userId'] ==2]).index)
df_2
```

```python
#Now we are creating a new dataframe which has all the similar users and their rated movies

final_df = df_2.merge(df, left_on='userId', right_on='userId', how='left')
#printing dataframe 
final_df
```

```python
#calculating the similarity score by multiplying user similarity and movie rating
final_df['score'] = final_df['similarity']*final_df['rating']
final_df
```

```python
#dataframe for all the movies which are already watched by targeted user i.e, userId = 2
watched_df = df[df['userId'] == 2]
#printing watched dataframe of user id 
watched_df
```

```python
#Removing all the movies that have already been watched by targeted user (userId = 2). 
#So, that watched movie by the targeted user won't be recommended again!

cond = final_df['movieId'].isin(watched_df['movieId'])
final_df.drop(final_df[cond].index, inplace = True) 
```

```python
recommended_df = final_df.sort_values(by = 'score', ascending = False)['title'].head(10)
recommended_df = recommended_df.reset_index()
del recommended_df['index']
```
